```python
from pathlib import Path
import pandas as pd
from ydata_profiling import ProfileReport
RAW = Path("data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
PROCESSED = Path("data/processed/telco_churn_clean.csv")
if PROCESSED.exists():
    df = pd.read_csv(PROCESSED)
else:
    df = pd.read_csv(RAW)
    df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
    df = df.dropna(subset=["TotalCharges"])
```

```python
profile = ProfileReport(df, title="Telco Churn YData Profile", explorative=True)
Path("notebooks/reports").mkdir(parents=True, exist_ok=True)
profile.to_file("notebooks/reports/telco_churn_ydata_profile.html")
```
